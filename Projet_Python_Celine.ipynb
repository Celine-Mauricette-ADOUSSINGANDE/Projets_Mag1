# **I- Introduction**

## **1-Presentation of the Topic**

The French government has implemented numerous measures aimed at ensuring the health and well-being of individuals (French or foreign) in France. Its high level of protection through the entire health insurance system makes the French healthcare system one of the most efficient and inclusive in the world. Indeed, the Health Insurance system sustainably protects everyone's health in their personal or professional lives by providing care regardless of their resources, situation, or health condition. Cancers (such as skin cancer, breast cancer, lung cancer, etc.) are among the diseases covered by health insurance, posing a major public health challenge in France and worldwide. They represent a significant portion of healthcare expenditures.

According to Ameli (2023), cancer is a disease caused by initially normal cells whose programming becomes disrupted and transforms them. These cells multiply and produce other cells, termed "abnormal," which proliferate anarchically and excessively. These disrupted cells eventually form a mass known as a malignant tumor, i.e., cancerous. Most cancers take several years to develop and can occur at any age, typically being diagnosed in individuals aged 60 and older (GENTESIDE, 2019). According to the National Cancer Institute (2024), cancer remains one of the leading causes of mortality in France, primarily affecting elderly individuals due to cellular aging. The number of new cancer cases for the year 2023 in mainland France is estimated at 433,136, with 245,610 in men and 187,526 in women. This number increases each year. Furthermore, all regions of France do not experience cancers equally. Beyond raw numbers, a detailed analysis of statistics published by the French Public Health Agency (Santé publique France, SPF) reveals significant disparities in both new cases (incidence) and cancer mortality depending on where one lives (D. Mascret, 2019).

Understanding the trends in healthcare expenditures related to cancer, as well as the factors associated with these expenditures, is crucial for developing effective health policies and managing medical resources.

This study aims, firstly, to analyze the temporal evolution of healthcare expenditures reimbursed by the health insurance system for cancer treatments in France from 2015 to 2021. Secondly, it aims to assess the impact of demographic characteristics, such as age, gender, and region, on reimbursed cancer expenses by analyzing variations between demographic groups. Finally, it aims to compare the levels of reimbursed expenses between regions and departments in France to determine regional and departmental variations and identify regions with the highest expense levels.

It's important to note that in our study, healthcare expenditures reimbursed by health insurance encompass the following:

- Outpatient care (including physician services, nursing or physiotherapy care, medication, laboratory tests, transportation, etc.);
- Hospitalizations in public or private healthcare facilities;
- Cash benefits (including daily allowances).

Our study focuses solely on healthcare expenditures reimbursed by health insurance for all beneficiaries affected by cancer. These expenditures may be partial or total reimbursements.

## **2- Presentation of Databases**

In order to achieve our objectives, we utilized two databases related to pathologies from 2015 to 2021, available on the Data Pathologies website (https://data.ameli.fr/pages/data-pathologies/):

- athologies : effectif de patients par pathologie, sexe, classe d'âge et territoire (département, région).
These data provide information on the number of patients treated by all health insurance schemes. They are available by pathology, chronic treatment, or episode of care, broken down by gender, age group, region, and department.

These data provide information on reimbursed expenses by all health insurance schemes allocated to each pathology, chronic treatment, or episode of care.

These data are produced by the National Health Insurance Fund (Cnam). It utilizes the population mapping of pathologies and expenses of the Health Insurance system to conduct numerous studies and produce data on pathologies and Health Insurance expenses.

The population mapping includes all beneficiaries of compulsory health insurance, regardless of their affiliation scheme:
- Those who received at least one service during the year (physician services, nursing or physiotherapy care, medication, laboratory tests, transportation, etc.);
- and/or those who stayed at least once in a public or private healthcare facility during the year (hospitalizations in medicine, surgery, obstetrics, psychiatry, follow-up care and rehabilitation, outpatient visits, or home hospitalization).

In 2021, it includes 68.7 million beneficiaries of all health insurance schemes who accessed reimbursed healthcare services.

```python
# #Aggregation of data by region and calculation of the total number of cancer cases for each region
df_cancers_par_regions = df_cancers_final.groupby('region')['Ntop'].size().reset_index()

# Merge geographical data with aggregated data by regionn
regions_cancers = regions.merge(df_cancers_par_regions, left_on='INSEE_REG', right_on='region', how='left')

# Calculate the total number of cancer cases
total_cancers = regions_cancers['Ntop'].sum()

# Calculate the percentage of cancer cases for each region
regions_cancers['percentage'] = (regions_cancers['Ntop'] / total_cancers) * 100

# Create an empty map
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

# Display the map with regions
regions_cancers.plot(column='Ntop', cmap='Reds', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# Add region names with larger labels and percentage as annotation
for idx, row in regions_cancers.iterrows():
    plt.annotate(text=f"{row['percentage']:.2f}%", xy=(row['geometry'].centroid.x, row['geometry'].centroid.y), color='white', fontsize=8, ha='center')
    plt.annotate(text=row['NOM'], xy=(row['geometry'].centroid.x, row['geometry'].centroid.y), color='black', fontsize=8, ha='center')
plt.title('Cancers par région en pourcentage')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

The observed cases of cancer are much higher in regions such as **Auvergne-Rhône-Alpes**, **Occitanie**, **Nouvelle-Aquitaine**, **Grand Est**, and **Île-de-France**. This indicates that the number of cancer cases varies from one region to another.

# **III- Conclusion**
The healthcare system in France is based on the missions of the Health Insurance. This institution, with its multiple branches, enables every insured person to benefit from reimbursement of their healthcare expenses related to illness, accidents, disability, or professional risks (Juliette F., 2023). In our study, we examined healthcare expenses related to cancer in France, analyzing data on reimbursements made by Health Insurance for the treatment of cancer patients. Our analysis revealed several significant trends and regional differences in healthcare expenses associated with cancer. Thus, based on data collected from 2015 to 2021 on expenses reimbursed by health insurance for beneficiaries with cancer across the entire French territory, we observed not only an increase in the number of patients being treated over time but also regional variations. Additionally, reimbursed expenses increase over the years but vary from one region to another, and therefore from one department to another. We also observed that the age of the patient is a determining factor because as age increases, reimbursed expenses also increase. These expenses are not influenced by gender and are particularly higher for patients aged **50 to 79**. This implies that the case of cancer becomes more critical and treatment requires higher expenses. It is therefore important to take these factors into account in the analysis of cancer treatment costs. This confirms our first two hypotheses, namely **the healthcare expenses reimbursed by the Health Insurance for cancer treatments increase over time due to the number of cancer cases per year** and **Reimbursed expenses vary depending on the demographic characteristics of beneficiaries, such as age, region, and department, due to differences in cancer incidence**. Furthermore, expenses reimbursed per patient are more often of a partial type than total. However, more than half (**66.7 %**) of the total amount of reimbursed expenses over the period is of a total type. This suggests that although the majority of reimbursed expenses are partial, total reimbursed expenses are costlier for the health insurance system than partial ones. This observation confirms our third hypothesis, which states **the difference in reimbursed expenses across regions and departments depends on the type of reimbursement (total or partial))**. These expenses represent a significant portion of the state's expenditure..

It's important to acknowledge the limitations of this study. Firstly, the data used only come from patient demographics and expenses reimbursed by the Health Insurance for cancer treatments, which may not reflect all costs associated with cancer treatments, such as patients' out-of-pocket expenses or indirect costs. Additionally, we did not include all observed cancer cases as we did not consider lines with missing values and lines that could potentially bias our analysis. We also did not directly consider the indication of cancer incidence or prevalence in our study.

Furthermore, this analysis focuses solely on expenses reimbursed by health insurance for cancer and does not consider other important aspects of cancer management, such as prevention, early detection, and patient support. Finally, while we have identified interesting trends, further studies are needed to better understand the underlying factors behind regional variations in cancer-related healthcare expenses and to formulate appropriate policy and clinical recommendations.

Finally, this study underscores the importance of continuing to monitor and evaluate reimbursed expenses related to cancer patient care to enhance the management of cancer patients in France. Future research could focus on identifying the determinants of regional variations in cancer patient numbers and therefore regional variations in reimbursed cancer-related expenses, as well as on the prevalence of cancers and according to different types of expenses (primary care services (including physician services, nursing or physiotherapy care, medication, laboratory tests, transportation, etc.), hospitalizations in public or private healthcare facilities, cash benefits (including daily allowances)).
